## Cargar las librerias necesarias

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Graficas

```python
#Coordenadas de la linea.
x = np.array([0,1,2,3,4])
y = np.array([0,2,4,6,8])

#Reescalando la grafica.
plt.figure(figsize = (5,3.5), dpi = 100)

#Aspecto de la linea.
#plt.plot(x,y, label = "2x", color = "red", linewidth = 3, marker = ".", markersize = 10, markeredgecolor = "blue", linestyle = "--")

#Con shortcut notation.
plt.plot(x,y, ".--r", label = "2x", markersize = 10, markeredgecolor = "blue") #Lo mismo pero mas cortito.
plt.title("Nightmare", fontdict = {"fontname" : "Times New Roman" , "fontsize" : 20})

#Nombre de los ejes.
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

#Numeros que aparecen en los ejes.
plt.xticks(np.array([0,1,2,3,4]))
#plt.yticks(np.array([0,2,4,6,8,10]))

#Grafica dos.
x2 = np.arange(0,4.5,0.5)

#Parte de la grafica como una linea.
plt.plot(x2[:5],x2[:5]**2, label = "x^2")
#Parte de la grafica como linea punteada.
plt.plot(x2[4:],x2[4:]**2, "--")

plt.legend()

#Guardar la grafica en el directorio donde esta mi notebook.
plt.savefig("nightmare.png", dpi = 300)

plt.show()
```

```python
#Coordenadas:
x = np.arange(0,4*np.pi,0.1) #Empezamos en cero, acabamos en 4pi y avanzamos 0.1 radianes.
y = np.sin(x) #Funcion de seno.

plt.figure(figsize = (5,3.5), dpi = 100)

plt.plot(x,y, "m", label = "sin(x)")
plt.title("Caca", fontdict = {"fontname" : "Times New Roman" , "fontsize" : 20})
plt.legend()
plt.show()


```
